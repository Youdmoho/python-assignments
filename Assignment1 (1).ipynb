```python
# Task 1
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

```python
# a) Filter out invalid transactions
# Create an empty list to store valid transactions
valid_transactions = []

# Loop through each transaction in the original list
for transaction in transactions:
    
    # Check if the transaction is valid
    # Condition: price must be greater than 0 AND quantity must be >= 0
    if transaction["price"] > 0 and transaction["quantity"] >= 0:
        
        # If valid, add it to the new list
        valid_transactions.append(transaction)
```

```python
# b) Compute total = price * quantity
# Loop through each valid transaction
for transaction in valid_transactions:
    
    # Calculate total value for the transaction
    total = transaction["price"] * transaction["quantity"]
    
    # Add the total as a new key-value pair in the dictionary
    transaction["total"] = total
```

```python
# c) Group by region and calculate revenue
# Create an empty dictionary to store revenue per region
revenue_by_region = {}

# Loop through each valid transaction
for transaction in valid_transactions:
    
    # Get region and total from the transaction
    region = transaction["region"]
    total = transaction["total"]
    
    # If the region already exists in the dictionary, add to it
    if region in revenue_by_region:
        revenue_by_region[region] += total
    else:
        # Otherwise, create a new entry for that region
        revenue_by_region[region] = total

# Print the revenue by region
print("Revenue by region:", revenue_by_region)
```

```python
# d) Find region with highest revenue
# Find the region with the maximum revenue
highest_region = max(revenue_by_region, key=revenue_by_region.get)

# Get the corresponding revenue value
highest_revenue = revenue_by_region[highest_region]

# Print the result
print("The region with the highest total revenue is", highest_region, "with", highest_revenue)
```

```python
# Task 2 
# Create a list of daily sales values
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

# Define a function that summarises sales data
def summarise_sales(sales_values):
    
    # Calculate the total sales
    total_sales = sum(sales_values)
    
    # Calculate the average sales and round to 2 decimals
    average_sales = round(total_sales / len(sales_values), 2)
    
    # Find the highest sales value and convert its position to day number starting from 1
    best_day = sales_values.index(max(sales_values)) + 1
    
    # Find the lowest sales value and convert its position to day number starting from 1
    worst_day = sales_values.index(min(sales_values)) + 1
    
    # Create a counter for days with sales above average
    above_average_count = 0
    
    # Loop through the sales values
    for value in sales_values:
        
        # Increase the counter if the value is above average
        if value > average_sales:
            above_average_count += 1
    
    # Return the results as a dictionary
    return {
        "total": total_sales,
        "average": average_sales,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }
```

```python
# Call the function and store the result
sales_summary = summarise_sales(sales)

# Print the returned dictionary
print(sales_summary)
```

```python
# Task 3
# Create a list of student records
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

```python
# a) Create a list of students who passed
# Create an empty list to store names of passing students
passing = []

# Loop through each student in the list
for student in students:
    
    # Check if the student's grade is 50 or higher
    if student["grade"] >= 50:
        
        # If yes, add the student's name to the list
        passing.append(student["name"])

# Print the result
print("Passing students:", passing)
```

```python
# b) Print a status message for each student
# Loop through each student in the list
for student in students:
    
    # Store name and grade in variables (makes code easier to read)
    name = student["name"]
    grade = student["grade"]
    
    # Check the grade and print the correct message
    if grade >= 80:
        print(name + ": Excellent")
    
    elif grade >= 50:
        print(name + ": Pass")
    
    else:
        print(name + ": Fail")
```

```python
# c) Count students with attendance below 0.80
# Create a counter variable and set it to 0
low_attendance_count = 0

# Loop through each student in the list
for student in students:
    
    # Check if the student's attendance is below 0.80
    if student["attendance"] < 0.80:
        
        # If yes, increase the counter by 1
        low_attendance_count += 1

# Print the result
print("Students with attendance below 0.80:", low_attendance_count)
```

```python
# d) Find the highest grade using a loop
# Assume the first student is the top student initially
top_student = students[0]

# Loop through each student
for student in students:
    
    # If current student has a higher grade, update top_student
    if student["grade"] > top_student["grade"]:
        top_student = student

# Print the result
print("Top student (loop):", top_student["name"], top_student["grade"])
```

```python
# e) Find the highest grade using max()
# Use max() with a key function to compare students by grade
top_student = max(students, key=lambda student: student["grade"])

# Print the student's name and grade
print("Top student:", top_student["name"], top_student["grade"])
```

```python
# Task 4

# Define the Portfolio class
class Portfolio:
    
    # Constructor: initializes owner and an empty dictionary for holdings
    def __init__(self, owner):
        self.owner = owner                    # Store the owner's name
        self.holdings = {}                    # Dictionary: ticker -> {shares, price_per_share}


    # a) Method to add or update an asset
    def add_asset(self, ticker, shares, price_per_share):
        
        # Check if the ticker already exists in the portfolio
        if ticker in self.holdings:
            
            # Update the number of shares
            self.holdings[ticker]["shares"] = shares
            
            # Update the price per share
            self.holdings[ticker]["price_per_share"] = price_per_share
        
        else:
            # If ticker does not exist, create a new entry
            self.holdings[ticker] = {
                "shares": shares,                  # Store number of shares
                "price_per_share": price_per_share # Store price per share
            }


    # b) Method to calculate total portfolio value
    def total_value(self):
        
        total = 0  # Initialize total value
        
        # Loop through all assets in the holdings dictionary
        for asset in self.holdings.values():
            
            # Calculate asset value (shares × price) and add to total
            total += asset["shares"] * asset["price_per_share"]
        
        # Return the total portfolio value
        return total


    # c) Method to find the most valuable asset
    def most_valuable(self):
        
        top_ticker = None     # Store ticker with highest value
        max_value = 0         # Store highest value found
        
        # Loop through all assets (ticker and data)
        for ticker, asset in self.holdings.items():
            
            # Calculate total value of current asset
            value = asset["shares"] * asset["price_per_share"]
            
            # Check if current asset has higher value
            if value > max_value:
                max_value = value        # Update highest value
                top_ticker = ticker      # Update ticker
        
        # Return the ticker of the most valuable asset
        return top_ticker


    # d) Method to return a readable summary of the portfolio
    def __str__(self):
        
        # Return formatted string with key information
        return (
            "Owner: " + self.owner +                                   # Owner name
            ", Number of assets: " + str(len(self.holdings)) +          # Number of assets
            ", Total value: " + str(self.total_value())                 # Total portfolio value
        )

# Testing the class
# Create a portfolio instance for Alice
portfolio = Portfolio("Alice")

# Add at least three assets
portfolio.add_asset("AAPL", 10, 180)
portfolio.add_asset("TSLA", 5, 250)
portfolio.add_asset("MSFT", 8, 320)

# Print portfolio summary (uses __str__)
print(portfolio)

# Print total portfolio value
print("Total value:", portfolio.total_value())

# Print most valuable asset
print("Most valuable asset:", portfolio.most_valuable())
```

```python
# Task 5 
# Import datetime from the datetime module
from datetime import datetime

# Create a list of orders as tuples
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Create an empty list to store parsed date objects
order_dates = []

# Loop through each order in the list
for order in orders:
    
    # Get the date string from the tuple
    date_string = order[0]
    
    # Convert the string to a datetime.date object
    parsed_date = datetime.strptime(date_string, "%Y-%m-%d").date()
    
    # Add the parsed date to the list
    order_dates.append(parsed_date)

# Find the earliest order date
earliest_date = min(order_dates)

# Find the latest order date
latest_date = max(order_dates)

# Calculate the number of days between the earliest and latest date
days_difference = (latest_date - earliest_date).days

# Print the result
print("Earliest order date:", earliest_date)
print("Latest order date:", latest_date)
print("Number of days between earliest and latest order:", days_difference)
```

```python
#Task 6
# Import NumPy
import numpy as np

# Create the revenue array (rows = years, columns = quarters)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])
```

```python
# a) Total annual revenue (row sums)

# Sum each row (axis=1 means across columns)
annual_revenue = np.sum(revenue, axis=1)

print("Annual revenue per year:", annual_revenue)

```

```python
# b) Mean quarterly revenue (column means)

# Mean of each column (axis=0 means across rows)
quarterly_mean = np.mean(revenue, axis=0)

print("Mean quarterly revenue:", quarterly_mean)
```

```python
# c) Apply growth adjustment to 2024

# Growth factors for each quarter
growth = np.array([1.05, 1.03, 1.08, 1.02])

# Apply broadcasting to the last row (2024)
adjusted_2024 = revenue[2] * growth

print("Adjusted 2024 revenue:", adjusted_2024)
```

```python
# d) Values greater than 250

# Boolean indexing to find values > 250
high_values = revenue[revenue > 250]

print("Revenue values above 250:", high_values)
```

```python
# Import pandas
import pandas as pd

# Create the DataFrame
data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
# a) Add salary_per_year_exp column

# Calculate salary divided by years of experience and round
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round()

print("DataFrame with new column:")
print(df)
```

```python
# b) Sort by performance_score (descending)

# Sort and select top 3 employees
top_3 = df.sort_values(by="performance_score", ascending=False).head(3)

print("\nTop 3 employees by performance:")
print(top_3)
```

```python
# c) Select Finance employees with score > 4.0

# Use .loc with conditions
finance_high = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]

print("\nFinance employees with high performance:")
print(finance_high)
```

```python
# d) Rename column years_exp → experience_years

# Rename column
df = df.rename(columns={"years_exp": "experience_years"})

print("\nDataFrame after renaming column:")
print(df)
```

```python
# e) Select salary and performance for first 4 employees

# Use .iloc to select rows and columns by position
selected_data = df.iloc[0:4][["salary", "performance_score"]]

print("\nSelected salary and performance (first 4 employees):")
print(selected_data)
```

```python
# Task 8
# Import NumPy and pandas
import numpy as np
import pandas as pd

# Generate the data
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

```python
# a) Create the DataFrame and set date as index

# Create a DataFrame from the generated data
traffic_df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

# Set the date column as the index
traffic_df = traffic_df.set_index("date")

# Print the DataFrame
print("Website traffic DataFrame:")
print(traffic_df)
```

```python
# b) Create conversions and daily_revenue columns

# Calculate conversions and round to nearest integer
traffic_df["conversions"] = (traffic_df["visitors"] * traffic_df["conversion_rate"]).round()

# Calculate daily revenue
traffic_df["daily_revenue"] = traffic_df["conversions"] * traffic_df["avg_order_value"]

# Print the updated DataFrame
print("\nDataFrame with new columns:")
print(traffic_df)
```
